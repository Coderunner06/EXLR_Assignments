# DBSCAN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime = pd.read_csv("crime_data.csv")
crime
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
# Normalize using standard scalar fit transform to dataset
crime_norm=StandardScaler().fit_transform(crime)
crime_norm
```

```python
model=DBSCAN(eps=1,min_samples=3)
model.fit(crime_norm)

```

```python
model.labels_
```

```python
crime['cluster']=model.labels_
```

```python
crime
```

```python
crime.cluster.value_counts()
```

```python
crime.groupby('cluster').agg(['mean']).reset_index()
```

# DBSCAN Clustering method created five clusters of the given crime data. The output explains that states with larger population are generally have higher crime rate however a few mid size states have much lower crime rate compared to the set of smaller towns. 

# Kmeans

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("crime_data.csv")
df
```

```python
df.info()
```

```python
x=df.iloc[:,1:].values

#feature scaling
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
```

```python
#calculating wcss to select optimum number of clusters
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#from elbow chart it is clear we should go for  either 2 or 4 clusters.
kmeans=KMeans(n_clusters=4,random_state=1)
kmeans.fit(x)
df['cluster']=kmeans.labels_
```
